```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import plotly.express as px
```

```python
#Mengimpor Data
data = pd.read_csv("D:/cs_studentss.csv",  sep = ";")
data.head()
```

```python
#Memeriksa data apakah ada missing value
np.sum(data.isnull())
```

```python
#Mengecek Data Outlier
plt.figure(figsize=(10,6))
sns.boxplot(data=data)
plt.show()
```

Visualisasi Data

```python
#Diagram Batang
sns.countplot(data=data, x="GPA")
plt.title("Diagram batang jumlah Nilai IPK Mahasiswa", fontsize=15)
plt.xlabel("IPK", fontsize=12)
plt.ylabel("Jumlah", fontsize=12)
plt.xticks(rotation=45) 
plt.show()
```

```python
#Diagram Garis
plt.figure(figsize=(10, 6))  
plt.plot(data['GPA'], data['Python'], marker='o')

plt.title("Diagram Garis Kemampuan Python Mahasiswa Ilmu Komputer menurut Nilai IPK")
plt.xlabel("IPK")
plt.ylabel("Kemampuan Python")
plt.xticks(rotation=45)  
plt.legend()
plt.tight_layout()  
plt.show()
```

```python
#Diagram Lingkaran
data_counts = data['Gender'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(data_counts, labels=data_counts.index, autopct='%1.1f%%', colors=plt.get_cmap('tab10').colors, shadow=True)
plt.title('Diagram Lingkaran perbandingan banyak laki-laki dan perempuan Mahasiswa Ilmu Komputer')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(data['GPA'], kde=True)
plt.title('Distribusi GPA')
plt.xlabel('GPA')
plt.ylabel('Frekuensi')
plt.show()
```

```python
#Diagram Pencar
plt.figure(figsize=(6,4))
sns.scatterplot(x='Age', y='GPA', data=data, color='purple')
plt.title('Korelasi Usia dengan GPA')
plt.show()
```

```python
#Diagram Gelembung
plt.figure(figsize=(10, 6))
plt.scatter(data['GPA'], data['Projects'], s=data['Age']*50, color='grey', alpha=0.5)
plt.title('Bubble Plot GPA vs Jumlah Projek (Ukuran: Age)')
plt.xlabel('GPA')
plt.ylabel('Projek')
plt.show()
```

```python
# Diagram lollipop 
plt.figure(figsize=(10, 6))
plt.stem(data['Future Career'], data['Python'], basefmt=" ")
plt.title('Kemampuan Python berdasarkan Karir Impian')
plt.xlabel('Kemampuan Python')
plt.ylabel('Projek')
plt.show()
```

```python
# Diagram polar
angles = np.linspace(0, 2 * np.pi, len(data.columns[7:10]), endpoint=False).tolist()
angles += angles[:1]

plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)
for i, row in data.iterrows():
    values = data.loc[i, ['Python', 'SQL', 'Java']].values.flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, label=row['Name'])
    ax.fill(angles, values, alpha=0.1)

ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(['Python', 'SQL', 'Java'])
plt.legend(loc='upper right')
plt.title('Diagram Polar Kemampuan Programming Mahasiswa')
plt.show()
```

```python
# Plot keterampilan pemrograman
plt.figure(figsize=(10,6))

# Python Skill
plt.subplot(1, 3, 1)
sns.histplot(data['Python'], kde=True, color='blue')
plt.title('Skill Python')

# SQL Skill
plt.subplot(1, 3, 2)
sns.histplot(data['SQL'], kde=True, color='green')
plt.title('Skill SQL')

# Java Skill
plt.subplot(1, 3, 3)
sns.histplot(data['Java'], kde=True, color='red')
plt.title('Skill Java')

plt.tight_layout()
plt.show()
```
