```python
import pandas as pd
data = pd.read_csv("D:/cs_studentss.csv",  sep = ";")
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew
```

```python
#Visualisasi Bar Plot Gender
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', data=data)
plt.title('Distribusi Gender')
plt.show()
```

Dilihat dari hasil visualisasi, kolom "Gender" termasuk ke dalam distribusi kategori. Distribusi ini menunjukkan seberapa banyak mahasiswa laki-laki dibandingkan dengan perempuan.

```python
#Visualisasi Bar Plot Kemampuan Python, SQL, dan Java
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

#Bar Plot Kemampuan Python
sns.countplot(x='Python', data=data, ax=axes[0])
axes[0].set_title('Kemampuan Python')

#Bar Plot Kemampuan SQL
sns.countplot(x='SQL', data=data, ax=axes[1])
axes[1].set_title('Kemampuan SQL')

#Bar Plot Kemampuan Java
sns.countplot(x='Java', data=data, ax=axes[2])
axes[2].set_title('Kemampuan Java')

plt.tight_layout()
plt.show()
```

Dilihat dari hasil visualisasi, kolom "Python", "SQL", dan "Java" termasuk ke dalam distribusi kategori. Distribusi ini menunjukkan berapa banyak mahasiswa yang berada di setiap tingkatan keterampilan untuk setiap kemampuannya dalam Python, SQL, dan Java.

```python
# Visualisasi histogram Age dan GPA
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(data['Age'], bins=10, kde=True, ax=axes[0], color='green')
axes[0].set_title('Histogram Age')
sns.histplot(data['GPA'], bins=10, kde=True, ax=axes[1], color='blue')
axes[1].set_title('Histogram GPA')
plt.tight_layout()
plt.show()
```

Dilihat dari hasil visualisasi, kolom "Age" termasuk ke dalam distribusi skewed (positif skew) karena memiliki ekor yang panjang di sebelah kanan. Ini berarti banyak mahasiswa di dataset ini memiliki usia yang lebih muda, namun terdapat beberapa data mahasiswa dengan usia yang jauh lebih tua yang menyebabkan distribusi condong ke kanan.

Dilihat dari hasil visualisasi, kolom "GPA" termasuk ke dalam distribusi skewed (negatif skew) karena memiliki ekor yang lebih panjang di sisi kiri (nilai yang lebih rendah), yang berarti lebih banyak data di sisi kanan (nilai yang lebih tinggi) dari rata-rata. Hal itu menunjukkan bahwa sebagian besar mahasiswa di dataset ini memiliki GPA yang lebih tinggi, dengan beberapa siswa memiliki GPA yang lebih rendah.

```python
#Pemusatan Data
mean_age = data['Age'].mean()
median_age = data['Age'].median()
modus_age = data['Age'].mode()[0]

mean_gpa = data['GPA'].mean()
median_gpa = data['GPA'].median()
modus_gpa = data['GPA'].mode()[0]

print(f"Mean Age: {mean_age}, Median Age: {median_age}, Modus Age: {modus_age}")
print(f"Mean GPA: {mean_gpa}, Median GPA: {median_gpa}, Modus GPA: {modus_gpa}")
```

```python
#Penyebaran Data
range_age = data['Age'].max() - data['Age'].min()
variance_age = data['Age'].var()
std_dev_age = data['Age'].std()

range_gpa = data['GPA'].max() - data['GPA'].min()
variance_gpa = data['GPA'].var()
std_dev_gpa = data['GPA'].std()

print(f"Age Range: {range_age}, Variance: {variance_age}, Std Dev: {std_dev_age}")
print(f"GPA Range: {range_gpa}, Variance: {variance_gpa}, Std Dev: {std_dev_gpa}")
```

```python
#Analisis Skewness
skew_age = skew(data['Age'])
skew_gpa = skew(data['GPA'])

print(f"Skewness of Age: {skew_age}")
print(f"Skewness of GPA: {skew_gpa}")
```

- Skewness positif (> 0):
Distribusi condong ke kanan. Artinya, sebagian besar data berada di sisi kiri (nilai kecil) dengan ekor panjang di sebelah kanan.
- Skewness negatif (< 0):
Distribusi condong ke kiri. Artinya, sebagian besar data berada di sisi kanan (nilai besar) dengan ekor panjang di sebelah kiri.
